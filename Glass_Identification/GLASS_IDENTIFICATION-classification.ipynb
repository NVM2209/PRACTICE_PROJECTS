The study of classification of types of glass was motivated by criminological investigation. At the scene of the crime, the glass left can be used as evidence...if it is correctly identified!


#### TYPE OF GLASS :

- There are 7 types of glasses grouped in 5 groups:-

 - 1 building_windows_float_processed
   
 - 2 building_windows_non_float_processed

  - 4 vehicle_windows_non_float_processed (none in this database)
   
  - 5 containers
   
  - 6 tableware
   
  - 7 headlamps

- This data set contain 214 instances, each corresponding to a glass/piece of glass. Each instance is described by 11 attributes.

#### PROBLEM IDENTIFICATION -
Identification of the glass at a crime scene can prove to be very useful in providing evidence in investigations and forensic purpose. Also, since glasses are used in various industries to manufacture different types of items, identifying the type beforehand from its components (oxide content) can save cost, time and effort.

#### AIM:- To predict the type of glass used from various components.

- Glass type

#### 1) RANDOM FOREST CLASSIFIER

```python
rf = RandomForestClassifier()
```

```python
# Finding out the best Random state for our model .

#split data, using train_test_split


maxAccu = 0            # Max Accuracy
maxRS = 0              # Best random state for which maximum accuracy is achieved.

for i in range(1,100):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=i)
    rf.fit(x_train,y_train)
    
    pred_test = rf.predict(x_test)
    acc_rf = accuracy_score(y_test,pred_test)
    
    print("Testing Accuracy: ",acc_rf,"at random state", i)
    
    if acc_rf > maxAccu:
        maxAccu = acc_rf
        maxRS = i
        
print('\n')
print("Max r2 score is ", maxAccu*100, "on Random state", maxRS)
```

#### Applying the best random state value in Random Forest Classifier

```python
#Splitting the data into training and testing Data.

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state = maxRS)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# training the data
rf.fit(x_train,y_train)
predrf = rf.predict(x_train)

rf_report = classification_report(y_train,predrf)

print("\n=======================Train Result===========================")
print(f"Accuracy score:{accuracy_score(y_train,predrf)*100:.2f}%")

print("_______________________________________________________________")
print(f"CLASSIFICATION REPORT:\n{rf_report}")
print("________________________________________________________________")
print(f" Confusion Matrix:\n{confusion_matrix(y_train,predrf)}\n")
```

```python
#testing the data

rf.fit(x_train,y_train)  
rf.score(x_train,y_train)
predrf = rf.predict(x_test)  # predicted values 

sc1 = metrics.accuracy_score(y_test,predrf)
print("\n=======================Test Result===========================")
print(f"Accuracy score:{accuracy_score(y_test,predrf)*100:.2f}%")
print("_______________________________________________________________")
print(f"CLASSIFICATION REPORT:\n",(classification_report(y_test,predrf)))
print("________________________________________________________________")
print(f" Confusion Matrix:\n{confusion_matrix(y_test,predrf)}\n")
```

#### CROSS VALIDATION OF RANDOM FOREST CLASSIFIER.

```python
#CV score of Random forest classifier-

score = cross_val_score(rf,x,y,cv=5)
print(score)
print(score.mean())
print("Accuracy score :", accuracy_score(y_test,predrf))
print(f"CV Score of RF:{cross_val_score(rf,x,y,cv=5).mean()*100:.2f}%")
print('\n')
print('The difference between accuracy score and Cross Validation score is:',accuracy_score(y_test,predrf)-score.mean())
```
