#### DATASET INFORMATION -

--This dataset is available on : https://github.com/dsrscientist/dataset4/blob/main/medical_cost_insurance.csv

Health insurance is a type of insurance that covers all medical expenses that arise due to an illness. These expenses could be related to hospitalisation costs, cost of medicines or doctor consultation fees. The main purpose of medical insurance is to receive the best medical care without any strain on your finances. Health insurance plans offer protection against high medical costs. It covers hospitalization expenses, day care procedures, domiciliary expenses, and ambulance charges, besides many others. 

Based on certain input features such as age , bmi,Sex,no of dependents ,smoker ,region,  medical cost insurance is calculated .
This dataset contains both numerical and categorical data.

This Dataset includes over 1338 instances and  7 attributes.

#### Problem Identification-
To study the accuracy and accurately predict insurance costs.

#### Attribute Information-

 -	Age: age of primary beneficiary
 -	Sex: insurance contractor gender, female, male
 -	BMI: Body mass index, providing an understanding of body, weights that are relatively high or low relative to height, objective index of body weight (kg / m ^ 2) using the ratio of height to weight, ideally 18.5 to 24.9.
 -	Children: Number of children covered by health insurance / Number of dependents
 -	Smoker: Smoking
 -	Region: the beneficiary's residential area in the US, northeast, southeast, southwest, northwest.
 -	Charges: Individual medical costs billed by health insurance

#### Checking Outliers

#### Removing the Outliers by IQR method

```python
Q1=data_copy.quantile(0.25)
Q3=data_copy.quantile(0.75)

IQR=Q3-Q1

df=data_copy[~((data_copy <(Q1-1.5*IQR))|(data_copy>(Q3+1.5*IQR))).any(axis=1)]

print('percentage of loss of data by IQR is :',(data_copy.shape[0]-df.shape[0])/data_copy.shape[0])
```

### MODEL BUILDING-

```python
# Models:
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.dummy import DummyRegressor
from sklearn.ensemble import RandomForestRegressor

# for cross validation
from sklearn.model_selection import cross_val_score,RandomizedSearchCV,GridSearchCV

# Matrics for Evaluation
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,accuracy_score,confusion_matrix
```

```python
#split data, using train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print('Size of X_train : ', x_train.shape)
print('Size of y_train : ', y_train.shape)
print('Size of X_test : ', x_test.shape)
print('Size of Y_test : ', y_test.shape)
```

### 1) Linear Regression

```python
# Training method

lm = LinearRegression()
lm.fit(x_train,y_train)
lm.score(x_train,y_train)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
#testing method

pred_train=lm.predict(x_train)
print(r2_score(y_train,pred_train))

pred_test=lm.predict(x_test)
print(r2_score(y_test,pred_test))
```

```python
cv_linearreg = cross_val_score(estimator = lm, X = x, y = y, cv = 5)

y_pred_linearreg_train = lm.predict(x_train)
r2_score_linearreg_train = r2_score(y_train, y_pred_linearreg_train)

y_pred_linearreg_test = lm.predict(x_test)
r2_score_linearreg_test = r2_score(y_test, y_pred_linearreg_test)

rmse = (np.sqrt(mean_squared_error(y_test, y_pred_linearreg_test)))

print('CV Linear Regression : {0:.3f}'.format(cv_linearreg.mean()))
print('R2_score (train) : {0:.3f}'.format(r2_score_linearreg_train))
print('R2_score (test) : {0:.3f}'.format(r2_score_linearreg_test))
print('RMSE : {0:.3f}'.format(rmse))
print('Mean absolute error :',mean_absolute_error(y_test,y_pred_linearreg_test))
print('Mean squared error :',mean_squared_error(y_test,y_pred_linearreg_test)) 
```

### 2) Support Vector Machine (Regression)

```python
svr=SVR()
svr.fit(x_train,y_train)
```

```python
ypred=svr.predict(x_test)
ypred
print('mean squared error :',np.sqrt(mean_squared_error(y_test,ypred)))
```

```python
kernellist = ['linear','poly','rbf']

for i in kernellist:
    svr = SVR(kernel = i)
    svr.fit(x_train,y_train)
    
    print ("SCORE OF:",i,"=",svr.score(x_train,y_train))
    
    svr.predict(x_test)
    print('mean squared error of :',i,np.sqrt(mean_squared_error(y_test,ypred)))
    
rmse_svr=np.sqrt(mean_squared_error(y_test,ypred))
rmse_svr
```

### Regularization

#### 3) Lasso Regularization

```python
ls=Lasso()
parameters={'alpha':[0.0001,0.001,0.01,0.1,1,10],
    'random_state':list(range(0,10))}

re=GridSearchCV(ls,parameters)
re.fit(x_train,y_train)
print(re.best_params_)
```

```python
#final model training
ls=Lasso(alpha=0.001,random_state=0)
ls.fit(x_train,y_train)
ls_score = ls.score(x_train,y_train)
print(ls.score(x_train,y_train)*100)
```

```python
predlasso=ls.predict(x_test)
print(mean_squared_error(y_test,predlasso))
rmse_lasso=np.sqrt(mean_squared_error(y_test,predlasso))
print('RMSE:',np.sqrt(mean_squared_error(y_test,predlasso)))
```

#### 4) Ridge Regularization

```python
rd = Ridge(alpha=.0001, random_state=42)
rd.fit(x_train, y_train)
print(rd.score(x_train,y_train))

cv_ridge = cross_val_score(estimator = rd, X = x, y = y, cv = 5)


y_pred_ridge_train = rd.predict(x_train)
r2_score_ridge_train = r2_score(y_train, y_pred_ridge_train)

y_pred_ridge_test = rd.predict(x_test)
r2_score_ridge_test = r2_score(y_test, y_pred_ridge_test)

rmse_ridge = (np.sqrt(mean_squared_error(y_test, y_pred_ridge_test)))
print('CV : {0:.3f}'.format(cv_ridge.mean()))
print('R2 score (train) : {0:.3f}'.format(r2_score_ridge_train))
print('R2 score (test) : {0:.3f}'.format(r2_score_ridge_test))
print('RMSE : {0:.3f}'.format(rmse_ridge))
```

#### 5)RandomForest Regressor

```python
parameters = {
    'criterion':['mse','mae'],
    'max_features':['auto','sqrt','log2']
}

rf=RandomForestRegressor()
clf=GridSearchCV(rf,parameters)
```

```python
clf.fit(x_train,y_train)
print(clf.best_params_)
```

```python
rf=RandomForestRegressor(criterion="mae",max_features="sqrt")
rf.fit(x_train,y_train)
rf.score(x_train,y_train)
pred_decision=rf.predict(x_test)

rfs=r2_score(y_test,pred_decision)
print('R2_score:',rfs*100)

rmse_rf = np.sqrt(mean_squared_error(y_test,pred_decision))

rfscore=cross_val_score(rf,x,y,cv=5)
rfc=rfscore.mean()
print('Cross Val Score:',rfc*100)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x=y_test,y=pred_decision,color='r')
plt.plot(y_test,y_test,color='b')

plt.xlabel('Actual Charges',fontsize=14)
plt.ylabel('Predicted charges',fontsize=14)
plt.title('Random Forest Regressor',fontsize=18)

plt.show()
```
