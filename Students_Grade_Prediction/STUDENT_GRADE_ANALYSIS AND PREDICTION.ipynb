#### MODEL BUILDING

```python
# Models:
from sklearn.model_selection import train_test_split,GridSearchCV

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso,Ridge
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.linear_model import SGDRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn import metrics

# for cross validation
from sklearn.model_selection import cross_val_score   

#Matrics for Evaluation:
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

##### Selecting the best random state 

#### 1)RANDOM FOREST REGRESSOR

```python
maxAcc = 0
maxRS = 0

for i in range(1,100):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.30,random_state=i)
    rf = RandomForestRegressor()
    rf.fit(x_train,y_train)
    predrf = rf.predict(x_test)
    acc = r2_score(y_test,predrf)
    if acc > maxAcc:
        maxAcc = acc
        maxRS = i
    print("R2 Score: ",acc,"at random state", i)
print("The Maximum r2 score is ", maxAcc*100, "on Random state", maxRS)
    
```

#### Applying the best random state value 

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=10)
```

```python
rf.fit(x_train,y_train)
predtrain = rf.predict(x_train)
predrf = rf.predict(x_test)
trainscore = r2_score(y_train,predtrain)
testscore = r2_score(y_test,predrf)

print("Training accuracy:",trainscore*100)

print("Testing accuracy:",testscore*100)
```

```python
# Plotting Predicted vs Actual CGPA
plt.figure(figsize=(8,6))
plt.scatter(y_test,predrf,color ='r')
plt.plot(y_test,y_test,color ='b')
plt.xlabel('Actual CGPA',fontsize = 14)
plt.ylabel('Predicted CGPA',fontsize =14)
plt.title('Random Forest Regression',fontsize = 18)
plt.show()
```

```python
cv_score=cross_val_score(ls,x,y,cv=5)
cv_mean=cv_score.mean()
cv_mean*100
```

```python
#applying all the model techniques together with the hep of for loop-

rf = RandomForestRegressor()
LR = LinearRegression()
GBR = GradientBoostingRegressor()
DTR = DecisionTreeRegressor()
ABR = AdaBoostRegressor()
KNR = KNeighborsRegressor()
svr = SVR()
BR = BaggingRegressor()
SGR = SGDRegressor()



Model = [rf,LR,GBR,DTR,ABR,KNR,svr,BR,SGR]

for m in Model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predmod = m.predict(x_test)
    R2score = r2_score(y_test,predmod)
    print("R2_score of", m,"is:",R2score*100)
    print("Mean absolute error: ", mean_absolute_error(y_test,predmod))
    print("Mean squared error: ", mean_squared_error(y_test,predmod))
    print("Root Mean absolute error: ",np.sqrt(mean_squared_error(y_test,predmod)))
    print('*************************************************************')
```

#### CROSS VALIDATION OF ALL THE MODELS 

```python
# checking the cross validation score 

cr1 = cross_val_score(rf,x,y, cv= 5)
cr2 = cross_val_score(LR,x,y, cv= 5)
cr3 = cross_val_score(GBR,x,y, cv= 5)
cr4 = cross_val_score(DTR,x,y, cv= 5)
cr5 = cross_val_score(ABR,x,y, cv= 5)
cr6 = cross_val_score(KNR,x,y, cv= 5)
cr7 = cross_val_score(svr,x,y, cv= 5)
cr8 = cross_val_score(BR,x,y, cv= 5)
cr9 = cross_val_score(SGR,x,y, cv= 5)



print('CV score of Random Forest Regressor:',cr1.mean()*100)
print('CV score of Linear Regression:',cr2.mean()*100)
print('CV score of Gradient Boost Regressor:',cr3.mean()*100)
print('CV score of Decision Tree Regressor:',cr4.mean()*100)
print('CV score of AdaBoostRegressor:',cr5.mean()*100)
print('CV score of K Neighbors Regressor:',cr6.mean()*100)
print('CV score of Support Vector Regressor:',cr7.mean()*100)
print('CV score of Bagging Regressor:',cr8.mean()*100)
print('CV score of SGDRegressor:',cr9.mean()*100)
```
